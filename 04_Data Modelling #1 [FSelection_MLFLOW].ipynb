```python
import mlflow
from mlflow.tracking import MlflowClient
```

```python
client = MlflowClient()
experiment = client.get_experiment_by_name("Feature Selection")
exp_id=experiment.experiment_id
data_=mlflow.search_runs([exp_id], order_by=["metrics.f1_micro DESC","metrics.log_loss ASC"])

rekap_FS={}
## For Catboost
rekap_FS['catboost-wo_artificial']=data_.loc[data_['params.algorithm']=='catboost',].loc[data_['params.strategy']=='no artificial'].iloc[0,:]
rekap_FS['catboost-w_artificial']=data_.loc[data_['params.algorithm']=='catboost',].loc[data_['params.strategy']!='no artificial'].iloc[0,:]

## For LightGBM
rekap_FS['lightgbm-wo_artificial']=data_.loc[data_['params.algorithm']=='lightgbm',].loc[data_['params.strategy']=='no artificial'].iloc[0,:]
rekap_FS['lightgbm-w_artificial']=data_.loc[data_['params.algorithm']=='lightgbm',].loc[data_['params.strategy']!='no artificial'].iloc[0,:]


## For Random Forest
rekap_FS['randomforest-wo_artificial']=data_.loc[data_['params.algorithm']=='randomforest',].loc[data_['params.strategy']=='no artificial'].iloc[0,:]
rekap_FS['randomforest-w_artificial']=data_.loc[data_['params.algorithm']=='randomforest',].loc[data_['params.strategy']!='no artificial'].iloc[0,:]
```

```python
with open('ml_output/04_05_modeling/feature_selection/fs.pickle', 'wb') as handle:
    pickle.dump(rekap_FS, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
